```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
#cost_data_low = pd.read_csv('cost_data_high.csv')
slowdown_data = pd.read_csv('data/slow_down_lr_curve.csv')


# Display the first few rows of the dataframe
print(slowdown_data.head())
```

```python
# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(slowdown_data.index,slowdown_data)

# Adding titles and labels
plt.title('slowdown')
plt.xlabel('Index')
plt.ylabel('Cost Data')

# Show the plot
plt.grid(True)
plt.show()
```
