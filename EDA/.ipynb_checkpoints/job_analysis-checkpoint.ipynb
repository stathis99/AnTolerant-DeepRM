```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Data/deeprm_jobs.csv')
```

```python
# Filter out data points where Size1, Size2, and JobLength are all 0
filtered_data = data[(data['Size1'] != 0) | (data['Size2'] != 0) | (data['# JobLength'] != 0)]

# Extract filtered values
JobLength = filtered_data["# JobLength"]
Size1 = filtered_data["Size1"]
Size2 = filtered_data["Size2"]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot histograms with automatically determined bins
axs[0].hist(JobLength, color='skyblue', edgecolor='black')
axs[0].set_title('JobLength Distribution')
axs[0].set_xlabel('JobLength')
axs[0].set_ylabel('Frequency')

axs[1].hist(Size1, color='salmon', edgecolor='black')
axs[1].set_title('Size1 Distribution')
axs[1].set_xlabel('Size1')
axs[1].set_ylabel('Frequency')

axs[2].hist(Size2, color='lightgreen', edgecolor='black')
axs[2].set_title('Size2 Distribution')
axs[2].set_xlabel('Size2')
axs[2].set_ylabel('Frequency')

# Show plot
plt.tight_layout()
plt.show()
```
