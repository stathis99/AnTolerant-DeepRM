```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_log = pd.read_csv('Data/jobs_lognorm.csv')
```

```python
data_log
```

```python
# Filter out data points where Size1, Size2, and JobLength are all 0
filtered_data = data_log[(data_log['Size1'] != 0) | (data_log['Size2'] != 0) | (data_log['# JobLength'] != 0)]

# Extract filtered values
JobLength = filtered_data["# JobLength"]
Size1 = filtered_data["Size1"]
Size2 = filtered_data["Size2"]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))  # Adjust figsize for desired dimensions

# Plot histograms with automatic binning
axs[0].hist(JobLength, color='skyblue', edgecolor='black')
axs[1].hist(Size1, color='salmon', edgecolor='black')
axs[2].hist(Size2, color='lightgreen', edgecolor='black')

# Set labels and titles
axs[0].set_title('Distribution of Job Length', fontsize=12)
axs[0].set_xlabel('Job Length', fontsize=10)
axs[0].set_ylabel('Frequency', fontsize=10)
axs[1].set_title('Distribution of CPU requirements', fontsize=12)
axs[1].set_xlabel('Size1', fontsize=10)
axs[1].set_ylabel('Frequency', fontsize=10)
axs[2].set_title('Distribution of Memory requirements', fontsize=12)
axs[2].set_xlabel('Size2', fontsize=10)
axs[2].set_ylabel('Frequency', fontsize=10)

# Adjust layout for readability
plt.tight_layout()

# Remove unnecessary grid lines (optional)
for ax in axs.flat:
    ax.grid(False)

# Display the plot
plt.show()
```
