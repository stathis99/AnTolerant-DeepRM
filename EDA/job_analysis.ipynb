```python
simu_len = 150
time = np.arange(simu_len)
arrival_rate = 0.5 + 0.5 * np.sin(2 * np.pi * time / (simu_len / 24))  # Varying sinusoidal arrival rate


x = np.zeros(simu_len, dtype=int)
for i in range(simu_len):
    x[i] = np.random.poisson(arrival_rate[i]+1)

binary_data = np.where(x == 0, 0, 1)

# Generate x values from 0 to 499
pltdata = np.arange(len(binary_data))

# Plot the data
plt.plot(pltdata, binary_data)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Binary Value')
plt.title('Binary Plot of Data')

# Show the plot
plt.show()
```

```python
num_ones = np.count_nonzero(binary_data)

# Calculate the percentage of 1s
percentage_ones = (num_ones / len(binary_data)) * 100

print("Percentage of 1s:", percentage_ones, "%")
```
