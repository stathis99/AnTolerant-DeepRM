```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming data_log is your DataFrame

# Filter out data points where Size1, Size2, and JobLength are all 0
filtered_data = data_log[(data_log['Size1'] != 0) | (data_log['Size2'] != 0) | (data_log['# JobLength'] != 0)]

# Extract filtered values
JobLength = filtered_data["# JobLength"]
Size1 = filtered_data["Size1"]
Size2 = filtered_data["Size2"]

fig, axs = plt.subplots(1, 2, figsize=(6, 3))  # Adjust figsize for desired dimensions

# Plot histograms with automatic binning
axs[0].hist(JobLength, color='#6e7f80', edgecolor='black')
axs[1].hist(Size1, color='#536872', edgecolor='black')
#axs[2].hist(Size2, color='#536872', edgecolor='black')

# Highlight areas where x > 20
highlight_color = 'red'
axs[0].axvspan(20, max(JobLength), color=highlight_color, alpha=0.3)
axs[1].axvspan(25, max(Size1), color=highlight_color, alpha=0.3)
#axs[2].axvspan(25, max(Size2), color=highlight_color, alpha=0.3)
# Set labels and titles
#axs[0].set_title('Job Length', fontsize=12)
axs[0].set_xlabel('Job Length', fontsize=10)
axs[0].set_ylabel('Frequency', fontsize=10)
#axs[1].set_title('CPU Requirements', fontsize=12)
axs[1].set_xlabel('CPU and Memory Requirements', fontsize=10)
axs[1].set_ylabel('Frequency', fontsize=10)
#axs[2].set_title('Memory Requirements', fontsize=12)
#axs[2].set_xlabel('Memory Requirements', fontsize=10)
#axs[2].set_ylabel('Frequency', fontsize=10)

# Adjust layout for readability
plt.tight_layout()

# Remove unnecessary grid lines (optional)
for ax in axs.flat:
    ax.grid(False)
    
plt.savefig('histograms.png', format='png')

# Display the plot
plt.show()
```

```python
simu_len = 150
time = np.arange(simu_len)
arrival_rate = 0.5 + 0.5 * np.sin(2 * np.pi * time / (simu_len / 24))  # Varying sinusoidal arrival rate


x = np.zeros(simu_len, dtype=int)
for i in range(simu_len):
    x[i] = np.random.poisson(arrival_rate[i]+1)

binary_data = np.where(x == 0, 0, 1)

# Generate x values from 0 to 499
pltdata = np.arange(len(binary_data))

# Plot the data
plt.plot(pltdata, binary_data)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Binary Value')
plt.title('Binary Plot of Data')

# Show the plot
plt.show()
```

```python
num_ones = np.count_nonzero(binary_data)

# Calculate the percentage of 1s
percentage_ones = (num_ones / len(binary_data)) * 100

print("Percentage of 1s:", percentage_ones, "%")
```
